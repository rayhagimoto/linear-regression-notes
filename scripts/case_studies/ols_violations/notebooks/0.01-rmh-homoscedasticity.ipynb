```python
from violations import HomoscedasticityStudy
import matplotlib.pyplot as plt
import numpy as np
```

```python
study = HomoscedasticityStudy()
study.simulate()
```

```python
def logspace_hist(x, n=100, **kwargs):
    x = np.array(x).flatten()
    bins = np.geomspace(np.clip(x.min()), x.max(), n)
    _ = plt.hist(x, bins, **kwargs)


logspace_hist(study.ols_betas, color='C0')
logspace_hist(study.wls_betas, color='C1')
plt.axvline(2.0, 0, 500, linestyle='dashed', color='red')
plt.xscale('log')
```
